## Eliminación de variables - Análisis de variables

### Correlación de Pearson

**Correlación**: una medida del grado de interdependencia entre variables.

**Causalidad**: la relación entre causa y efecto entre dos variables.

Es importante saber la diferencia entre estos dos y que la correlación no implica causalidad. Determinar la correlación es mucho más simple que la causalidad determinante ya que la causalidad puede requerir experimentación independiente.

La correlación de Pearson mide la dependencia lineal entre dos variables X e Y.

El coeficiente resultante es un valor entre -1 y 1 inclusive, donde:

+ 1: correlación lineal positiva total.
+ 0: Sin correlación lineal, las dos variables probablemente no se afectan entre sí.
+ -1: correlación lineal negativa total.

Correlación de Pearson es el método predeterminado de la función "corr". Como antes, podemos calcular la correlación de Pearson de las variables 'int64' o 'float64'.

### Valor p

¿Qué es este valor P? El valor P es el valor de probabilidad de que la correlación entre estas dos variables sea estadísticamente significativa. Normalmente, elegimos un nivel de significancia de 0.05, lo que significa que estamos 95% seguros de que la correlación entre las variables es significativa.

Por convención, cuando el

+ El valor p es <0.001: decimos que hay pruebas sólidas de que la correlación es significativa.
+ El valor p es <0.05: existe evidencia moderada de que la correlación es significativa.
+ El valor p es <0.1: existe evidencia débil de que la correlación es significativa.
+ El valor p es > 0.1: no hay evidencia de que la correlación sea significativa.

### Chi-cuadrado

Analizaremos mediante la técnica de chi-cuadrado la importancia de las **variables categóricas**.

```python
columns_categ = ['TIPOUSUARIO','FECHA_REGISTRO_WEEKDAY','FECHA_REGISTRO_MONTH','CANAL_REGISTRO','TIPOEMAIL','BONDAD_EMAIL',
                 'TIPO_TELF','IP_CONTINENTE','IP_COLOMBIA','IP_COUNTRY_PIB','USU_TIPO','USU_TAMANIO','USU_ESTADO','USU_SECTOR_ECONOMICO','FECHACONSUMO_WEEKDAY',
                 'FECHACONSUMO_MONTH','1A_DESCGRUPOPROD', '1A_EMPCONSUL_EST','1A_EMPCONSUL_SECTOR','IND_CLIENTE']
```

```python
for var in columns_categ:
    if var != 'IND_CLIENTE':

        print('')
        print('Análisis de Chi-cuadrado de independencia de la variable: ', var)
        print('')
        
        print('TABLA DE CONTINGENCIA')
        print('')
        contigency = pd.crosstab(df[var], df['IND_CLIENTE'])
        print(contigency)
        print('')

        # Chi-square test of independence.
        c, p, dof, expected = chi2_contingency(contigency)
        print('El valor de Chi-Cuadrado: ', c)
        print('El valor p: ', p)
        
        if contigency.shape[0] <= 2:
            print('La matriz de casos experados: ')
            print(pd.DataFrame(expected, columns = ['-','-'], index = ['-','-']))
        
        print('')
        if p < 0.05:
            print('podemos CONSIDERAR la variable ', var, ' como estadísticamente significativa con la vraible objetivo!!')

        else:
            print('podemos DESCARTAR la variable ', var, ' por no ser estadísticamente significativa con la vraible objetivo!!')

        print('')
        print('#############################################################################')
        print('')
```

-----------------

#### **Conclusiones**:
+ Tras este análisis, podemos afirmar que todas las variables categóricas aportan información y son estadísticamente significativas.

A esto se concluye observando los valores de p del calculo de la chi cuadrado. (Para todos ellos << 0.5)

### Coeficiente de Cramer

Analizaremos la correlación de las variables categóricas en base al coeficiente de Cramer, utilizando chi cuadrado.

**Definimos la función**

```python
def cramers_corrected_stat(matrix):
    
    chi2 = chi2_contingency(matrix)[0]
    n = matrix.sum().sum()
    phi2 = chi2/n
    
    r,k = matrix.shape
    phi2corr = max(0, phi2 - ((k - 1)*(r - 1))/(n - 1))
    
    rcorr = r - ((r - 1)**2)/(n - 1)
    kcorr = k - ((k - 1)**2)/(n - 1)
    
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))
```

Aplicamos la función

```python
for i in columns_categ:
    if (i != 'IND_CLIENTE'):
        
        print('Calculo del coeficiente de cramer para la variable ', i)
        print('')
        print('------------------------------------------------------')
        
        for j in columns_categ:
            if (j != 'IND_CLIENTE') & (i != j):
                
                matrix = pd.crosstab(df[j], df[i])
                
                if cramers_corrected_stat(matrix) > 0.5:
                    
                    print('Respecto con la variable: ', j)
                
                    print('')
                    print('El coeficiente de Cramer es: ', cramers_corrected_stat(matrix))
                    print('')
                    print('######################################################')
                
        print('######################################################')
```

-----------------

**Para interpretar los valores de Cramer, utilizaremos esta tabla**

| Coeff. Cramer | Descripción  
|---|:---:|
| > 0.5 | Alta correlación |
| 0.3 - 0.5 | Moderada correlación |
| 0.1 - 0.3 | Baja correlación |
| < 0.1 | Poca o ninguna correlación |
